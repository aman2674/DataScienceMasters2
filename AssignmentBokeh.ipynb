# Q1. How can you create a Bokeh plot using Python code?
**Ans :-**

```python
from bokeh.plotting import figure, show

# Define data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a figure object
p = figure(title="Simple Line Plot", x_axis_label='x', y_axis_label='y')

# Add a line glyph to the figure
p.line(x, y, legend_label="Line", line_width=2)

# Show the plot
show(p)
```

# Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.
**Ans :-**
Glyphs in Bokeh are visual markers that represent data points on a plot, such as lines, circles, squares, or bars. You can add glyphs to a Bokeh plot using various methods provided by the Bokeh library, such as the line(), circle(), square(), or bar() functions.

```python
from bokeh.plotting import figure, show

# Define data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a figure object
p = figure(title="Circle Glyph Example", x_axis_label='x', y_axis_label='y')

# Add a circle glyph to the figure
p.circle(x, y, legend_label="Circle", size=10, color="navy", alpha=0.5)

# Show the plot
show(p)
```

### Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?
**Ans :-**
You can customize the appearance of a Bokeh plot using various attributes and methods provided by the Bokeh library. 

```python
from bokeh.plotting import figure, show

# Define data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a figure object
p = figure(title="Customized Plot Example", x_axis_label='x', y_axis_label='y')

# Add a line glyph to the figure
p.line(x, y, legend_label="Line", line_width=2)

# Customize appearance
p.title.text_color = "navy"
p.title.text_font_size = "20px"
p.xaxis.axis_label_text_color = "red"
p.yaxis.axis_label_text_font_style = "italic"
p.legend.label_text_font = "times"

# Show the plot
show(p)
```

### Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in real time?
**Ans :-**
A Bokeh server is a Python server that allows you to create and serve interactive Bokeh plots. With a Bokeh server, you can create plots that update in real-time in response to user interactions or changes in data.

To use a Bokeh server, you need to define a Python script that creates the plot and specifies the interactions or updates. You then run the script using the Bokeh server command (bokeh serve) to start the server and serve the plot. Users can access the plot through a web browser, and the plot updates dynamically based on their interactions or changes in the data.

### Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?
**Ans :-**

```python
from flask import Flask, render_template
from bokeh.plotting import figure
from bokeh.embed import components

app = Flask(__name__)

@app.route('/')
def index():
    # Create Bokeh plot
    p = figure(title="Embedded Bokeh Plot")
    p.circle([1, 2, 3], [4, 5, 6])
    
    # Generate JavaScript and HTML components
    script, div = components(p)
    
    # Pass components to template
    return render_template('index.html', script=script, div=div)

if __name__ == '__main__':
    app.run(debug=True)
```
