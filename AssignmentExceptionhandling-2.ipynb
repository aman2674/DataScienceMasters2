### Q1. Explain why we have to use the Exception class while creating a Custom Exception.
Note: Here Exception class refers to the base class for all the exceptions.

**Ans:-** In object-oriented programming languages like Python, the Exception class serves as the base class for all exceptions. When creating custom exceptions, it is recommended to inherit from the Exception class or one of its subclasses. 
 - **Consistent Hierarchy:** By inheriting from the Exception class, your custom exception becomes part of the existing exception hierarchy in the language. 
- **Compatibility:** Inheriting from the Exception class ensures that your custom exception is compatible with the existing exception-handling mechanisms in the language.

### Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.
**Ans:-** The ArithmeticError class in Python is a base class for exceptions that occur during arithmetic operations. It serves as the parent class for several specific arithmetic-related exception classes.

```python
#ZeroDivisionError
try:
    result = 5 / 0  # Attempting to divide by zero
except ZeroDivisionError as e:
    print(f"Error: {e}")
```

```python
# OverflowError
import math

try:
    result = math.exp(1000)  # Attempting to calculate the exponential of a large number
except OverflowError as e:
    print(f"Error: {e}")
```

### Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
**Ans:-** The LookupError class in Python serves as the base class for exceptions that occur when trying to access an index or key that is not present in a sequence or mapping. 

```python
# KeyError:
# This exception is raised when attempting to access a dictionary key that does not exist.
my_dict = {'name': 'John', 'age': 30, 'city': 'New York'}

try:
    value = my_dict['gender']  # Attempting to access a key that does not exist
except KeyError as e:
    print(f"Error: {e}")
```

```python
# IndexError:
#This exception is raised when trying to access an index in a sequence (e.g., list, tuple) that is out of range.
my_list = [10, 20, 30, 40, 50]

try:
    element = my_list[10]  # Attempting to access an index that is out of range
except IndexError as e:
    print(f"Error: {e}")
```

### Q5. Explain ImportError. What is ModuleNotFoundError?
**Ans:-** 
ImportError is an exception class in Python that is raised when an import statement fails to find the specified module or when there is an issue during the import process.


One specific subclass of ImportError is ModuleNotFoundError. This subclass is raised when Python cannot locate the module specified in the import statement.


```python
try:
    import non_existent_module  # Attempting to import a module that does not exist
except ImportError as e:
    print(f"Error: {e}")
```

### Q6. List down some best practices for exception handling in python.
**Ans:-**
- Use Specific Exceptions: Catch specific exceptions rather than using a generic except block. 
- Avoid Bare except Blocks:

Avoid using bare except blocks as they catch all exceptions, including system exits and keyboard interrupts.
