### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
**Ans:-**
1. **Scatter Plot**: A scatter plot is used to visualize the relationship between two continuous variables. It helps in identifying patterns, trends, and correlations between the variables.

2. **Line Plot**: A line plot is used to visualize the relationship between two continuous variables over time or another continuous variable. It is useful for showing trends and changes in data over time.

3. **Bar Plot**: A bar plot is used to compare categorical data or to show the distribution of a categorical variable. It is helpful in visualizing the frequency or proportion of different categories.

4. **Box Plot**: A box plot (or box-and-whisker plot) is used to display the distribution of a continuous variable and to identify outliers. It provides information about the central tendency, variability, and skewness of the data.

5. **Histogram**: A histogram is used to visualize the distribution of a single continuous variable. It shows the frequency or count of data points within predefined bins or intervals.

### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.
**Ans:-**

```python
import seaborn as sns
import matplotlib.pyplot as plt

fmri_data = sns.load_dataset("fmri")
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region")
plt.title("FMRI Signal Over Time for Different Events and Regions")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.show()
```

### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
**Ans :-**

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic_data = sns.load_dataset("titanic")
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
sns.boxplot(data=titanic_data, x='pclass', y='age')
plt.title('Box Plot of Age by Passenger Class')

plt.subplot(1, 2, 2)
sns.boxplot(data=titanic_data, x='pclass', y='fare')
plt.title('Box Plot of Fare by Passenger Class')

plt.tight_layout()
plt.show()
```

### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.
**Ans :-**

```python
import seaborn as sns
import matplotlib.pyplot as plt

diamonds_data = sns.load_dataset("diamonds")
sns.histplot(data=diamonds_data, x='price', hue='cut', kde=True)
plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

### Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.
**Ans :-**

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris_data = sns.load_dataset("iris")
sns.pairplot(data=iris_data, hue="species")
plt.title("Pair Plot of Iris Dataset")
plt.show()
```

### Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
**Ans :-**


```python
import seaborn as sns
import matplotlib.pyplot as plt

flights_data = sns.load_dataset("flights")
flights_pivot = flights_data.pivot("month", "year", "passengers")
sns.heatmap(data=flights_pivot, cmap="YlGnBu")
plt.title("Passenger Counts by Month and Year")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```
